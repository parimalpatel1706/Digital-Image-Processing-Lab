```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("dip.png")
```

```python
type(img)
```

```python
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))# here we converting bgr image into rgb 
plt.title("Original image")
plt.axis("off")
```

```python
print("Filename: dip.png")
print("Image Shape:", img.shape)
print("Image Size (pixels):", img.size)
print("Data Type:",img.dtype)
```

```python
height, width, channels = img.shape
print("Width:", width)
print("Height:", height)
```

```python
cv2.imwrite("saved_dip.png",img)
```

```python
## Channels
B = img[:, :, 0]## here :(coln) show indexing 
G = img[:, :, 1]
R = img[:, :, 2]
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,4,1)## here image is divide into 4 columns and 1 channel here
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title("Original")
plt.axis("off")

plt.subplot(1,4,2)## here channel is 2 and here subplot is used for indexing
plt.imshow(R, cmap='gray')
plt.title("Red")
plt.axis("off")

plt.subplot(1,4,3)## here channel is 3
plt.imshow(G, cmap='gray')
plt.title("Green")
plt.axis("off")

plt.subplot(1,4,4)## here channel is 4
plt.imshow(B, cmap='gray')
plt.title("Blue")
plt.axis("off")
```

```python
## Pure image convert into gray
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray, cmap='gray')
plt.title("Grayscale Image")
plt.axis("off")
```

```python
### converting GRAY to RGB
gray_rgb = cv2.cvtColor(gray, cv2.COLOR_GRAY2RGB)
plt.imshow(gray_rgb)
plt.title("Gray to RGB")
plt.axis("off")
```

```python
indexed = (gray // 52).astype(np.uint8)
plt.imshow(indexed, cmap='gray')
plt.title("Indexed Image (Concept)")
plt.axis("off")
```

```python
bit_plane = (gray >> 7) &1

plt.imshow(bit_plane, cmap='gray')
plt.title("MSB Bit Plane")
plt.axis("off")
```
